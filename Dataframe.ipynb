```python
import polars as pl
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from pathlib import Path
import time

# ==========================================
# INICIO DEL CONTADOR
# ==========================================
inicio = time.perf_counter()

# ==========================================
# CREAR CARPETA DE SALIDA
# ==========================================
Path("graficas").mkdir(exist_ok=True)

# ==========================================
# CARGA DEL DATASET
# ==========================================
t_carga_inicio = time.perf_counter()

archivo_parquet = Path("dataset_completo.parquet")
archivo_csv = Path("dataset_completo.csv")

if archivo_parquet.exists():
    print("Cargando archivo Parquet...")
    df = pl.read_parquet(archivo_parquet)
else:
    print("Cargando archivo CSV...")
    df = pl.read_csv(
        archivo_csv,
        infer_schema_length=10000
    )

t_carga_fin = time.perf_counter()

# ==========================================
# DICCIONARIO DE ENTIDADES
# ==========================================
entidades = {
    1: "AGUASCALIENTES",
    2: "BAJA CALIFORNIA",
    3: "BAJA CALIFORNIA SUR",
    4: "CAMPECHE",
    5: "COAHUILA DE ZARAGOZA",
    6: "COLIMA",
    7: "CHIAPAS",
    8: "CHIHUAHUA",
    9: "CIUDAD DE MÉXICO",
    10: "DURANGO",
    11: "GUANAJUATO",
    12: "GUERRERO",
    13: "HIDALGO",
    14: "JALISCO",
    15: "MÉXICO",
    16: "MICHOACÁN DE OCAMPO",
    17: "MORELOS",
    18: "NAYARIT",
    19: "NUEVO LEÓN",
    20: "OAXACA",
    21: "PUEBLA",
    22: "QUERÉTARO",
    23: "QUINTANA ROO",
    24: "SAN LUIS POTOSÍ",
    25: "SINALOA",
    26: "SONORA",
    27: "TABASCO",
    28: "TAMAULIPAS",
    29: "TLAXCALA",
    30: "VERACRUZ DE IGNACIO DE LA LLAVE",
    31: "YUCATÁN",
    32: "ZACATECAS",
    36: "ESTADOS UNIDOS MEXICANOS",
    97: "NO APLICA",
    98: "SE IGNORA",
    99: "NO ESPECIFICADO"
}

# ==========================================
# REEMPLAZAR CÓDIGOS
# ==========================================
t_transformacion_inicio = time.perf_counter()

df = df.with_columns([
    pl.col("ENTIDAD_UM")
      .replace_strict(entidades, default="DESCONOCIDO")
      .alias("ENTIDAD_UM_NOMBRE"),

    pl.col("ENTIDAD_RES")
      .replace_strict(entidades, default="DESCONOCIDO")
      .alias("ENTIDAD_RES_NOMBRE")
])

t_transformacion_fin = time.perf_counter()

# ==========================================
# CREAR MATRIZ DE CONTINGENCIA
# ==========================================
t_matriz_inicio = time.perf_counter()

matriz = (
    df.group_by(
        ["ENTIDAD_RES_NOMBRE", "ENTIDAD_UM_NOMBRE"]
    )
    .len()
    .pivot(
        on="ENTIDAD_UM_NOMBRE",
        index="ENTIDAD_RES_NOMBRE",
        values="len"
    )
    .fill_null(0)
)

# Ordenar filas y columnas alfabéticamente
columnas_ordenadas = sorted(
    [c for c in matriz.columns if c != "ENTIDAD_RES_NOMBRE"]
)

matriz = matriz.select(
    ["ENTIDAD_RES_NOMBRE"] + columnas_ordenadas
).sort("ENTIDAD_RES_NOMBRE")

t_matriz_fin = time.perf_counter()

# ==========================================
# GUARDAR MATRIZ EN CSV
# ==========================================
matriz.write_csv(
    "graficas/matriz_ENTIDAD_RES_vs_ENTIDAD_UM.csv"
)

# ==========================================
# CONVERTIR A PANDAS PARA GRAFICAR
# ==========================================
matriz_pd = matriz.to_pandas()
matriz_pd = matriz_pd.set_index("ENTIDAD_RES_NOMBRE")

# ==========================================
# CREAR HEATMAP
# ==========================================
t_grafica_inicio = time.perf_counter()

fig, ax = plt.subplots(figsize=(18, 16))

imagen = ax.imshow(
    matriz_pd.values + 1,  # evita problemas con valores cero
    aspect="auto",
    cmap="inferno",
    norm=LogNorm()
)

ax.set_xticks(range(len(matriz_pd.columns)))
ax.set_xticklabels(
    matriz_pd.columns,
    rotation=90,
    fontsize=8
)

ax.set_yticks(range(len(matriz_pd.index)))
ax.set_yticklabels(
    matriz_pd.index,
    fontsize=8
)

ax.set_xlabel("ENTIDAD_UM")
ax.set_ylabel("ENTIDAD_RES")

ax.set_title(
    "Matriz de contingencia: ENTIDAD_RES vs ENTIDAD_UM"
)

# Líneas divisorias
ax.set_xticks(
    [x - 0.5 for x in range(1, len(matriz_pd.columns))],
    minor=True
)

ax.set_yticks(
    [y - 0.5 for y in range(1, len(matriz_pd.index))],
    minor=True
)

ax.grid(
    which="minor",
    color="white",
    linestyle="-",
    linewidth=0.2
)

cbar = plt.colorbar(imagen, ax=ax)
cbar.set_label("Número de registros (escala logarítmica)")

plt.tight_layout()

plt.savefig(
    "graficas/matriz_ENTIDAD_RES_vs_ENTIDAD_UM.png",
    dpi=300,
    bbox_inches="tight"
)

plt.close()

t_grafica_fin = time.perf_counter()

# ==========================================
# TIEMPOS DE EJECUCIÓN
# ==========================================
fin = time.perf_counter()

print("\n" + "=" * 60)
print("ARCHIVOS GENERADOS")
print("=" * 60)

print(
    "Tabla:    graficas/matriz_ENTIDAD_RES_vs_ENTIDAD_UM.csv"
)

print(
    "Gráfica:  graficas/matriz_ENTIDAD_RES_vs_ENTIDAD_UM.png"
)

print("\n" + "=" * 60)
print("TIEMPOS DE EJECUCIÓN")
print("=" * 60)

print(f"Carga del dataset:         {t_carga_fin - t_carga_inicio:.2f} s")
print(f"Transformación de datos:  {t_transformacion_fin - t_transformacion_inicio:.2f} s")
print(f"Creación de la matriz:    {t_matriz_fin - t_matriz_inicio:.2f} s")
print(f"Generación de gráfica:    {t_grafica_fin - t_grafica_inicio:.2f} s")
print("-" * 60)
print(f"Tiempo total:             {fin - inicio:.2f} s")
print("=" * 60)
```

```python
fuera = (
    df.filter(
        pl.col("ENTIDAD_RES").is_between(1, 32)
    )
    .with_columns(
        (pl.col("ENTIDAD_RES") != pl.col("ENTIDAD_UM"))
        .alias("FUERA")
    )
    .group_by("ENTIDAD_RES")
    .agg([
        pl.len().alias("TOTAL"),
        pl.col("FUERA").sum().alias("FUERA_TOTAL")
    ])
    .with_columns(
        (pl.col("FUERA_TOTAL") * 100 / pl.col("TOTAL"))
        .round(2)
        .alias("PORC_FUERA")
    )
    .rename({"ENTIDAD_RES": "CVE_ENT"})
    .with_columns(
        pl.col("CVE_ENT")
        .cast(pl.Utf8)
        .str.zfill(2)
    )
)

fuera.write_csv("Datos_QGis/QGIS_PORCENTAJE_FUERA.csv")
```
